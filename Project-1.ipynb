1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух словах и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
def education_parser(text:str)->str:
    """Finds out the level of education of the candidate

    Args:
        text (str): parsed text (format:<"Уровень образования etc">)

    Returns:
        str: category of education ("высшее","неоконченное высшее","среднее специальное","среднее") // can be None?
    """    
    text_list = text.split()
    if text_list[0] == 'Высшее':
        return 'высшее'
    if text_list[0] == 'Неоконченное':
        return 'неоконченное высшее'
    if text_list[1] == 'специальное':
        return 'среднее специальное'
    if text_list[0] == 'Среднее':
        return 'среднее'
 
 
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(education_parser)

print("Полученный столбец")
display(hh_data['Образование'].head())
print("Убедимся, что получили только 4 категории")
display(hh_data['Образование'].value_counts())
print("Убедимся, что число записей в признаке 'Образование' совпадает с числом записей в признаке 'Образование и Вуз':", end='')
print(hh_data['Образование'].shape[0] == hh_data['Образование и ВУЗ'].shape[0])

#delete column "Образование и ВУЗ"
hh_data.drop("Образование и ВУЗ", axis = 1, inplace = True) 
pass
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def months_parser(text:str)->int:
    """Extract experience in number of months

    Args:
        text (str): text with description of experience(format: "<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>")
    Returns:
        int: experience in number of months
    """    
    #ключевые слова
    month_words = ['месяц','месяцев','месяца']
    year_words = ['лет','год','года']
    if text == 'Не указано':
        return np.nan
    try:
        text_list = text.split()
        text_list = text_list[:7] # усечем список, не важны периоды работы в различных компаниях
    except AttributeError: # raise if text is "Не указано" 
        return np.nan 
    
    year=0
    month=0
    for i, item in enumerate(text_list):
        if item in month_words:
            month = int(text_list[i-1]) #сохраним месяц, если найдено ключевое слово
        if item in year_words:
            year = int(text_list[i-1]) #сохраним года, если найдено ключевое слово
    return int(12*year + month)


hh_data["Опыт работы (месяц)"] = hh_data['Опыт работы'].apply(months_parser)
hh_data.drop('Опыт работы', axis = 1, inplace = True)

#Чему равен медианный опыт работы (в месяцах) в нашей таблице?
answer = hh_data["Опыт работы (месяц)"].median()
print(f'Медианный опыт работы среди соискателей {answer} месяцев.')
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def city_parser(text:str)->str:
    """Extract city from the text

    Args:
        text (str): text to parse, format: <Город , (метро) , 
                                    готовность к переезду (города для переезда) , 
                                    готовность к командировкам>

    Returns:
        str: city of employee
    """     
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', \
                      'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', \
                      'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    city = text.split(',')[0].strip() # extract city from the text
    #categorize city
    if city == 'Москва' or city == 'Санкт-Петербург':
        return city
    if city in million_cities:
        return 'город-миллионник'
    return 'другие'


def relocation_parser(text:str)->bool:
    """Get information about readiness of employee to relocation

    Args:
        text (str): text to parse, format: <Город , (метро) , 
                                    готовность к переезду (города для переезда) , 
                                    готовность к командировкам>

    Returns:
        bool: True - ready, False - doesnt ready
    """    
    key_word = ['переезд', 'переезжать', 'переехать', 'переезду']
    text_list = text.split(',')
    #Текст без указания метро
    for partical in text_list[1].split():
        if partical.strip() == "не":
            return False
    #Текст с указанием метро
    for partical in text_list[1].split():
        if partical == "не":
            for word in text_list[1].split():
                if word.strip() in key_word:
                    return False
    for partical in text_list[2].split():
        if partical == "не":
            for word in text_list[2].split():
                if word.strip() in key_word:
                    return False
    return True
        

def trip_parser(text:str)->bool:
    """Get information about readiness of employee to business trip

    Args:
        text (str): text to parse, format <Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>

    Returns:
        bool: True - ready, False - doesnt say or doesnt ready
    """    
    key_word = ['командировкам', 'командир', 'команд']
    text_list = text.split(',')
    #string without metro
    for word in text_list[-1].split():
        if word.strip() in key_word:
            if 'не' not in text_list[-1].split():
                return True
    return False


#Выделение отдельных столбцов
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(city_parser)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(trip_parser)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(relocation_parser)

#Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
answer1 = round(hh_data['Город'].value_counts(normalize=True)['Санкт-Петербург']*100)
print(f'В Санкт-Петербурге живет {answer1}% соискателей.')

#Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
answer2 = round(100*hh_data[hh_data['Готовность к командировкам'] & hh_data['Готовность к переезду']].shape[0]/hh_data.shape[0])
print(f'Готовы к переездами и к командировкам одновременно {answer2}% соискателей.')

hh_data.drop('Город, переезд, командировки', axis = 1, inplace = True)
pass
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
#Найдем уникальные категории занятости
temp = hh_data['Занятость'].apply(lambda x:x.split(','))
occupations_list = []
for item in temp:
    for word in item:
        occupations_list.append(word.strip())
        
occupations_list = set(occupations_list)

#Найдем уникальные категории графика работы
temp = hh_data['График'].apply(lambda x:x.split(','))
timetable_list = []
for item in temp:
    for word in item:
        timetable_list.append(word.strip())        
timetable_list = set(timetable_list)

def text_to_list(text:str)->list:
    """Search in text search_word, function for "apply"

    Args:
        text (str): text in dataframe column
        search_word (str): word wich we search for hot key encoding

    Returns:
        bool: True - word in text, False - dont find word 
    """    
    text_list = text.split(',')
    clean_list = []
    for word in text_list:
        clean_list.append(word.strip())
    return clean_list

#Переберем категории и создадим новые колонки для колонки "График"
for category in timetable_list:
    hh_data[category] = hh_data['График'].apply(lambda x: True if category in text_to_list(x) else False) #не учитываю что хранятся слова с пробелами по краям
    
#Переберем категории и создадим новые колонки для колонки "Занятость"
for category in occupations_list:
    hh_data[category] = hh_data['Занятость'].apply(lambda x: True if category in text_to_list(x) else False)
    
#Удалим лишние колонки
hh_data.drop('Занятость', axis = 1, inplace=True)
hh_data.drop('График', axis = 1, inplace=True)

#Сколько людей ищут проектную работу и волонтёрство (в обоих столбцах стоит True)?
#Сколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)?
answer1 = hh_data[hh_data['волонтерство'] & hh_data['проектная работа']].shape[0]
answer2 = hh_data[hh_data['вахтовый метод'] & hh_data['гибкий график']].shape[0]
print(f'Хотят работать волонтерами и ищут проектную работу {answer1} соискателей.')
print(f'Хотят работать вахтовым методом с гибким графиком {answer2} соискателей.')
```

6. (2 балла) Наконец, мы добрались до самого главного и самого важного - признака заработной платы **"ЗП"**. 
В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:
* 30000 руб.
* 50000 грн.
* 550 USD

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
exchange_df = pd.read_csv('data/ExchangeRates.csv', sep=',')
display(exchange_df.head())

#Создадим словарь переименований
exchange_template = dict(zip(hh_data['ЗП'].apply(lambda x: x.split()[1]).value_counts().index,['руб.','KZT','USD','BYN','EUR','UAH','UZS','KGS','AZN']))

#1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
hh_data['дата'] = pd.to_datetime(hh_data['Обновление резюме']).dt.date
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме']).dt.date
exchange_df['дата'] = pd.to_datetime(exchange_df['date']).dt.date

#2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. 
# Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
hh_data['ЗП единиц'] = hh_data['ЗП'].apply(lambda x: float(x.split()[0].strip()))
hh_data['currency'] = hh_data['ЗП'].apply(lambda x: x.split()[1].strip())
hh_data['currency'] = hh_data['currency'].apply(lambda x: exchange_template[x])

#3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты 
# (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). 
# Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
hh_data = hh_data.merge(exchange_df[['currency','дата','proportion','close']],\
                            left_on = ['дата','currency'], \
                            right_on = ['дата', 'currency'], \
                            how = 'left')
hh_data['close'] = hh_data.apply(lambda row: 1 if row['currency'] == 'руб.' else row['close'], axis = 1) #заполняем значения close для руб. единицами
hh_data['proportion'] = hh_data.apply(lambda row: 1 if row['currency'] == 'руб.' else row['proportion'], axis = 1) #заполняем значения proportion для руб. единицами

#4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию 
# (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".
hh_data['ЗП руб.'] = hh_data['close']*hh_data['ЗП единиц']/hh_data['proportion'] #вычисляем ЗП в руб = close*единицы/proportion

#Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)?
answer = hh_data['ЗП руб.'].median()
print(f'Медианная желаемая зарплата соискателей равна {answer} руб.')

#Сбросим служебные столбцы и сверимся с итоговым dataframe.info() на сайте 
сolumns_to_drop = ['ЗП','close','ЗП единиц','currency','дата','proportion']
hh_data.drop(columns_to_drop, axis = 1, inplace = True)
display(hh_data.info())
```

Есть ли аномалии для признака возраста, какие значения вы бы причислили к их числу?
*Совет: постройте гистограмму и коробчатую диаграмму рядом.*

```python
# ваш код здесь
# добавим функцию, вычисляющую границы распределения
def iqr_bounds(data:pd.DataFrame, feature:str)->tuple:
    """Calculate iqr bounds with as 1.5 IQR

    Args:
        data (pd.DataFrame): dataframe to research
        feature (str): numeric column

    Returns:
        tuple: (lower bound, upper_bound)
    """    
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    return lower_bound, upper_bound


fig, axes = plt.subplots(nrows = 2, ncols= 1, figsize = (10,10))
# коробчатая диаграмма
age_box = sns.boxplot(
    hh_data,
    x = 'Возраст',
    ax = axes[0]
)
# гистограмма
age_hist = sns.histplot(
    hh_data,
    x = 'Возраст',
    bins = 100,
    ax = axes[1]
)
# добавим вертикальные линии внутри которых будут находится все соискатели, 
# и которые покажут минимальные и максимальный возраст соискателя
# добавим вертикальную линию, указывающую на моду распределения
mode_distribution = hh_data['Возраст'].mode()[0]
line1 = axes[0].axvline(hh_data['Возраст'].min(), c='red', linestyle='--')
line2 = axes[0].axvline(hh_data['Возраст'].max(), c='green', linestyle='--')
lower_bound = iqr_bounds(hh_data,'Возраст')[0] if iqr_bounds(hh_data,'Возраст')[0] > hh_data['Возраст'].min() else hh_data['Возраст'].min()
upper_bound = iqr_bounds(hh_data,'Возраст')[1]
line3 = axes[1].axvline(mode_distribution, c = 'red', linestyle = '--') #нарисуем линию моды
line4 = axes[1].axvline(lower_bound, c = 'gray', linestyle = '--') #нижняя граница распределения
line5 = axes[1].axvline(upper_bound, c = 'gray', linestyle = '--') #верхняя граница распределения
#Settings
axes[0].set_title('Коробчатая диаграмма распределения возраста ')
axes[0].set_xlabel('Возраст')
axes[0].set_xticks(list(axes[0].get_xticks()) + [hh_data['Возраст'].min(), hh_data['Возраст'].max()])
axes[0].legend([line1, line2], ['Min', 'Max'])
axes[1].set_title('Гистограмма распределения возраста ')
axes[1].set_xlabel('Возраст')
axes[1].set_ylabel('Количество человек в возрастной группе')
axes[1].set_xticks(list(axes[1].get_xticks()) + [mode_distribution,lower_bound,upper_bound])
axes[1].legend([line3,line4,line5], ['Mode','Lower bound','Upper bound'])
axes[1].grid()

print(f'Мода распределения возраста равна {mode_distribution} годам.')
```
